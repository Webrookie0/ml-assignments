```python
import pandas as pd

# Redefine X and feature columns based on available data
categorical_cols = ['State/UT']
numerical_cols = [
    '2020',
    '2021',
    '2022',
    'Mid-Year Projected Population (in Lakhs) (2022)',
    'Chargesheeting Rate (2022)'
]
X = df[categorical_cols + numerical_cols]

# Redefine target variables based on available data
y_target_column = 'Rate of Cognizable Crimes (IPC) (2022)'
y_class = pd.qcut(df[y_target_column], q=3, labels=["Low", "Medium", "High"])
y_reg = df[y_target_column]

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
        ("num", StandardScaler(), numerical_cols)
    ]
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_class,
    test_size=0.25,
    random_state=42,
    stratify=y_class
)
```

```python
k_values = [1, 3, 5, 7, 9]
accuracies = []

for k in k_values:
    clf = Pipeline(steps=[
        ("preprocess", preprocessor),
        ("knn", KNeighborsClassifier(n_neighbors=k))
    ])

    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)

    print(f"K = {k}, Accuracy = {acc:.3f}")
```

```python
plt.figure()
plt.plot(k_values, accuracies, marker="o")
plt.xlabel("K value")
plt.ylabel("Classification Accuracy")
plt.title("KNN Classification: Accuracy vs K")
plt.show()
```

```python
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X,
    y_reg,
    test_size=0.25,
    random_state=42
)
```

```python
rmse_values = []

for k in k_values:
    reg = Pipeline(steps=[
        ("preprocess", preprocessor),
        ("knn", KNeighborsRegressor(n_neighbors=k))
    ])

    reg.fit(X_train_r, y_train_r)
    y_pred_r = reg.predict(X_test_r)

    rmse = np.sqrt(mean_squared_error(y_test_r, y_pred_r))
    rmse_values.append(rmse)

    print(f"K = {k}, RMSE = {rmse:.2f}")
```
